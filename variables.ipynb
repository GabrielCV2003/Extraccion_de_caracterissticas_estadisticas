```python
#Cargar desde un archivo .xlsx sin indice
data = pd.read_csv('Mexico_City_Limpio(1).csv')
```

```python
data = data.drop('Unnamed: 0', axis=1)
data.head()
```

```python
data.info()
```

```python
freq_tbl(data)
```

**COLUMNA PRICE**

```python
#Calculamos el numero total de la poblacion
data['price'].info()
```

```python
n= 26582
```

```python
#Obtenemos el limite suerior y el limite inferior de la columna objetivo
Max=data['price'].max()
Min=data['price'].min()
Limites =[Min, Max]
Limites
```

```python
#Calculamos el r
R=Max- Min
R
```

```python
#Calculamos el numero de Intervalos e Clase "ni" aplicando la regla de Sturges
ni = 1+3.322*np.log10(n)
ni
```

```python
#Calculamos el ancho del intervalo "I"
i = R/ni
i
```

```python
#Categorizacion de variables
intervalos=np.linspace(Min, Max, 13)
intervalos
```

```python
#Creamos las categorias
categorias= ["0 - 226.8", "226.08 - 452.16", "452.16 - 678.25", "678.25 - 904.33", "904.33 - 1130.41", "1130.41 - 1356.5", "1356.5 - 1582.58", "1582.58 - 1808.66", "1808.66 - 2034.75", "2260.83 - 2486.91", "2486.91 - 2713", "2713"]
```

```python
#Finalmente creamos las categorias en la columna numerica
data['price']=pd.cut(x= data["price"], bins=intervalos, labels = categorias)
data['price']
```

```python
table25 = freq_tbl(data['price'])
table25
```

```python
table25 = freq_tbl(data['price'])
table25
```

```python
table25_1 = table25.drop(['percentage','cumulative_perc'], axis=1)
table25_1
```

```python
Filtro25 = table25_1[table25_1['frequency']>1]
Filtro25
```

```python
#Ajusto el inidice de mi dataframe
Filtro_index25 = Filtro25.set_index('price')
Filtro_index25
```

```python
Filtro_index25.plot(kind = 'bar', width=0.8, figsize=(10,4), color = "red")
plt.title('Precios')
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')
```

```python
#Realizamos grafico de dispercion del dataframe filtrado
table25.plot("frequency", "cumulative_perc", kind="scatter")
```

```python
Filtro_index25.plot(kind='area', figsize=(10,4), alpha = 0.8)
```

```python
Filtro_index25["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct= "%0.1f ")
```
